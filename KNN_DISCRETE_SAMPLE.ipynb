```python
import pandas as pd
```

```python
data = {
    'Height': [150, 155, 160, 161, 158],
    'Weight': [50, 55, 60, 59, 65],
    'Target': ['m', 'm', 'l', 'l', 'l']
}

df = pd.DataFrame(data)
df
```

```python
sample = [157, 54]
```

```python
#Calculating Distance
distances=[]
#print(df.iloc[2].tolist())
i=0
while i<len(df):
  j=0
  sum = 0
  while j<len(df.columns)-1:
    feat = (df.iloc[i].tolist()[j])
    smple = sample[j]
    sum = sum + (feat-smple)**2
    j=j+1
  distances.append(sum**(1/2))
  i=i+1

print(distances)
```

```python
df['dist'] = distances
df
```

```python
k=3
#Sorting DF
df = df.sort_values(by='dist', ascending=True)
df
```

```python
def argmax(a, b):
  if (a==b):
    return 1
  else:
    return 0

V = df['Target'].unique().tolist()
fin_vals = {}
for v in V:
  sum=0
  for row in df.head(3).itertuples():
    sum = sum + argmax(v, row.Target)
  fin_vals[v] = sum
print(fin_vals)
```

```python
max_key = max(fin_vals, key=fin_vals.get)
print("Final Class identified - ", max_key)
```
